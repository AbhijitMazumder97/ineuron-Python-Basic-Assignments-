Q17. How do you distinguish between copy.copy() and copy.deepcopy()?

Answer:17 
copy.copy() is used to make a duplicate copy of a mutable value like a list or a dictionary, whereas copy.deepcopy() function is used when the list to be copied contains lists inside it. copy.deepcopy() copies the inner lists as well.
