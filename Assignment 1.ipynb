Q9. 

What three functions can be used to get the integer, floating-point number, or string version of a value?

Answer:
The three functions can be used to get the integer, floating-point number, or string version of a value are:
    
int() - used to convert an integer stored in a variable (usually user input,i.e, input()) in the form of a string to integer form. Also used to round up a floating number. 

str() - used to convert values of other data types to a string form usually preferred during string concatenation operation.

float() - used to convert values of other data types into their decimal forms. It cannnot convert any string to float but values stored in variables (especially integers) can be converted to their decimal form.

Q10. 

Why does this expression cause an error? How can you fix it?

'I have eaten ' + 99 + ' burritos.'

Answer: 

Cannot concatenate strings and integers.



```python
'I have eaten' + 99 + 'burritos.'
```

```python
## In order to fix the above problem, we use typecasting i.e., we convert integer type 99 into a string form which helps in concatenation.
'I have eaten ' + str(99) + ' burritos.'
```
