Q10. How can you tell the difference between break and continue?

Answer:  

A 'break' statement is used within a loop. The program execution will exit the loop when a 'break' statement is executed.
Just like a 'break' statement 'continue' statement is also used inside a loop but unlike the 'break' statement, the program execution will jump back to the start of loop on execution of a continue statement and reevaluate the loop's condition.

Q12. Write a short program that prints the numbers 1 to 10 using a for loop. Then write an equivalent program that prints the numbers 1 to 10 using a while loop.

```python
## Using for loop....

for i in range(10):
    i=i+1
    print('number:',i)
```

```python
## Using while loop....

m=0
while m<10:
    m=m+1
    print('number:',m)
```
