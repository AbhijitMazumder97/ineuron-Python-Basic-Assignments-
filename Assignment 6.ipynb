**Q1. What are escape characters, and how do you use them?**

**Answer:1.**  
An ***escape character*** lets you use characters that are otherwise impossible to put into strings. An escape character consists of a backslash \ followed by the character you want to add to the string.  
*For example:*

```python
'This is Rahul's lunchbox.'
```

```python
'This is Rahul\'s lunchbox.'
```

**Q2. What do the escape characters n and t stand for?**

**Answer:2.**  
The following escape characters stand for:  
**\n** stands for 'new line'.  
**\t** stands for 'tab'.

**Q3. What is the way to include backslash characters in a string?**

**Answer:3.**  
Double backslash (**\\\**) represents backslash characters in a string.
