Q14. What is the purpose of the try clause? What is the purpose of the except clause?

Answer:  
The code that could potentially have an error is put in a try clause.If there is no error in the code then the program execution will not jump to the except clause and execute the entire code. Otherwise on encountering an error the program execution jumps to the except clause immediately and executes the code within. Except statement is called only when an error is detected in try clause.
