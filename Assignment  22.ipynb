# Q2. What is the result of the code, and explain?


>>> X = 'iNeuron'

>>> def func():
         X = 'NI!'
         


>>> func()

>>> print(X)


**Ans:2**

```python
X ='iNeuron'

def func():
    X = 'NI!'
```

```python
func()
```

```python
print(X)
```

***X = 'iNeuron' is a global variable whereas X = 'NI!' is a local variable. Local variable (i.e., X = 'NI!') is recognised only inside the function 'func()' whereas X='iNeuron' is recognised throughout the entire program. On calling func(), the value of the variable X(inside the func() scope) gets replaced by 'NI!' but the outside X variable remains intact. On calling print(X) the value of global variable 'X' is displayed.***

# Q5. What about this code—what’s the output, and why?


>>> X = 'iNeuron'

def nested():
print(X)
nested()


>>> func()
>>> X


**Ans:5**

```python
X = 'iNeuron'

def func():
    X = 'NI'
    def nested():
        print(X)
    nested()
```

```python
func()
```

***On calling the func(), value of variable X is replaced with 'NI' (inside the func() only), and then finally on calling the nested() the value of 'X' gets printed.***

```python
X
```

***Returns the value of variable 'X'(Global scope value).***

# Q6. How about this code: what is its output in Python 3, and explain?


>>> def func():
X = 'NI'
def nested():
nonlocal X
X = 'Spam'
nested()
print(X)


>>> func()

**Ans:6**

```python
def func():
    X = 'NI' # Enclosing scope starts
    def nested():
        nonlocal X #Local scope starts
        X = 'Spam' # Local scope ends 
    nested()
    print(X)  # Enclosing scope ends  
```

***By using nonlocal keyword we're able to modify the value of any variable inside enclosing scope from the local scope.***

```python
func()
```

***On calling func(), the variable 'X' is assigned a value 'NI' and nested() function gets called which is another function inside func(). Inside nested() due to the use nonlocal keyword, the enclosing/nonlocal scope variable's ('X') value gets changed permanently on assigning the new value 'Spam' inside the local scope. Therefore the calling func() will further call nested() and 'Spam' is returned on execution of the entire code.***
